```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('iris.csv')
print(data.head())
```

```python
selected_features = ['SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']
data = data[selected_features].dropna()
print(data)
```

```python
scaler = StandardScaler()
scaled_data = scaler.fit_transform(data)
```

```python
pca = PCA(n_oversamples=3)
principal_components = pca.fit_transform(scaled_data)
```

```python
pc_df = pd.DataFrame(data=principal_components,columns=['PC1','PC2','PC3'])
print(pc_df)
```

```python
explained_variance = pca.explained_variance_ratio_
plt.bar(['PC1', 'PC2','PC3'], explained_variance)
plt.xlabel('Principal Components')
plt.ylabel('Explained Variance Ratio')
plt.title('Explained Variance Ratio of Principal Components')
plt.show()
```

```python
data_with_pc = pd.concat([pc_df], axis=1)
print(data_with_pc.head())
```

```python
import seaborn as sns
sns.heatmap(data_with_pc.corr(),annot=True) 
```
